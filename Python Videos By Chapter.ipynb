### f-strings for output
* Python f-strings are used **inside print() statements**.
* They enable control of every character in the output.
* f-strings feature **placeholders** defined by braces **{ }**.
* A variable name inside a placeholder will display in the output.
* **Designators** can control **how** a variable displays in output.
* Designators control alignment, decimals, and field widths.

```python
# Intro to f-strings
YouTubeVideo('f_SClMaWiD0')
```

```python
# f-string program from video above

n = 64
rt2 = n ** 0.5      ## square root of n
rt3 = n ** (1/3)    ## cube root of n

# n is LEFT-aligned with 2 decimals in a 10-character column
# rt2 is CENTERED with 5 decimals in a 12-character column
# rt3 is RIGHT-aligned with 5 decimals in a 12-character column
    
print(f'|{n:<10.2f}|{rt2:^12.5f}|{rt3:>12.5f}|')
print(f'|{n:<10.2f}|{rt2:^12.5f}|{rt3:>12.5f}|')
print(f'|{n:<10.2f}|{rt2:^12.5f}|{rt3:>12.5f}|') 
```

## <font color=red>CHAPTER 3 - DECISION STRUCTURES</font>
* Programs can run one code block if a condition is **True** and another if it is **False**.
* Conditions are expressed with a **boolean expression** in an **if statement**.  
* If the expression is True, all **indented statements** in the block are executed.
* An **else block** can run a different code block if the condition is False.
* When more than two branches are needed, **elif** blocks are used.

```python
# good program from the video above
# prompt user for quantity
    
qty = int(input('Enter quantity of Nuggit meals as an integer '))

if qty >= 50:
    disc_rate = 0.25
elif qty >= 25:
    disc_rate = 0.15       
elif qty >= 10:
    disc_rate = 0.10
else:
    disc_rate = 0.00
    print('Ordering more Nuggit meals can get you a discount')        

cost = qty * 4.99 * (1-disc_rate)
print(f'Your discount for quantity is {disc_rate:.0%}')
print(f'Please pay ${cost:.2f}')
```

## <font color=red>CHAPTER 4 - LOOPS</font>
* Loops are code structures that **repeat** code blocks.
* Each loop cycle is called an **iteration**.

### There are **two loop types** in Python:
1. **for loops**
   * for loops are **count-controlled** loops.
   * for loops often use the **range()** function for counting.
1. **while loops**
   * while loops are **condition-controlled** loops.
   * while loops **iterate** as long as a condition is **True**.
   * An **endless loop** results if the condition never becomes **False**.

## <font color=red>CHAPTER 5 - FUNCTIONS</font>
* Python has many functions, but you can **make your own custom functions**.

### There are **two types** of custom functions.
1. **Void functions**
   * These execute one or more statements **without returning a value**.
   * Often, although not always, they print something.
1. **Value-returning functions**
   * These do return a value, **back** to the function that called them.
   * To be useful, the returned value **must be captured** in some way.
   * It could be **caught in a variable**, or used **as an argument** to another function.

## <font color=red>CHAPTER 7 - LISTS</font>
* A list is a group of values with one name.
* Think of a list as a variable with multiple "slots" for values.
* A Python list is **like an array** in other programming languages.

```python
nums = list(range(80,-5,-5))
for n in nums:
    print(n, end = ' ')
print()

dex = 0  # element indexing starts at zero
while dex < len(nums):
    print(nums[dex], end = ' ')
    dex += 1
print()

# display list backwards by index
# index of final element is length minus 1
n = len(nums) - 1
while n >= 0:
    print(nums[n], end = ' ')
    n -= 1
print()
    
# printing the name of a list gives "crude" output    
print(nums) # crude because brackets and commas display
```
